# Plots & Graphs


```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(100)
y=x/1

plt.title("Matplot Example:")
plt.xlabel("X-axis")
plt.ylabel("Y=axis")

plt.plot(x,y,linestyle='-',color='c',marker='o')
plt.show()
```

```python
#Sine and Cosine
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(0,3*np.pi,0.1)
ysin=np.sin(x)
ycos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,ysin)
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,ycos)
plt.title('cosine')

plt.subplots_adjust(hspace=0.3)
plt.suptitle('-------->> Sine and cosine <--------')
plt.show()
```

```python
import matplotlib.pyplot as plt

# dataset-1
x1 = [89, 43, 36, 36, 95, 10, 66, 34, 38, 20]

y1 = [21, 46, 3, 35, 67, 95, 53, 72, 58, 10]

# dataset2
x2 = [26, 29, 48, 64, 6, 5, 36, 66, 72, 40]

y2 = [26, 34, 90, 33, 38, 20, 56, 2, 47, 15]

plt.scatter(x1, y1, c ="indigo",
			linewidths = 1,
			marker ="s",
			edgecolor ="green",
			s = 50)

plt.scatter(x2, y2, c ="yellow",
			linewidths = 1,
			marker ="o",
			edgecolor ="red",
			s = 200)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
